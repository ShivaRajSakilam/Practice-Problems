# rank() method in pandas

```python
import pandas as pd
nba=pd.read_csv('nba.csv').dropna(how='all')
nba['Salary']=nba['Salary'].fillna(0).astype('int')
nba.head()
```

```python
#I want to decide on the basis of salary, that whose salary is highest comes at rank 1 and so on..
nba['Salary'].rank()#Computes numerical data ranks along axis.
# nba['Salary'][91]
nba['Salary'].rank(ascending=False).astype(int).head()
nba['Salary Rank']=nba['Salary'].rank(ascending=False).astype(int)#Decides Whether or not the elements should be ranked in ascending order
nba.sort_values(by='Salary', ascending=False)
# # nba.head()
```

# Query Methods in Pandas

```python
#Query methods only applys when all the column are strings and no space in between columns name

import pandas as pd
bond=pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head()
```

```python
bond.columns=[column_names.replace(" ",  "_") for column_names in bond.columns]#replacing spaces of column names with '_'
bond.head()
```

```python
bond.query('Actor=="Sean Connery"')#shows the dataframe satisfying the particular column condition
```

```python
bond.query("Director=='Terence Young'")#shows the dataframe satisfying the particular column condition
```

```python
bond.query("Actor != 'Roger Moore'")#shows the dataframe satisfying the particular column condition
```

```python
bond.query("Box_Office>600")#shows the dataframe satisfying the particular column condition
```

```python
bond.query("Bond_Actor_Salary>3")#shows the dataframe satisfying the particular column condition
```

```python
bond.query("Actor=='Roger Moore' and Director=='John Glen'")#shows the dataframe satisfying the particular columns condition
```
